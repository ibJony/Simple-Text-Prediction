```python
# Load my dataset here
def myDataset():
    
    batch_size = 32
    
    x = np.load(os.path.join('train_feats.npy')).astype(float)
    y = np.load(os.path.join('train_labels.npy'))
    #print(type(x[0]),x[1].shape)
    xTst= np.load(os.path.join('test_feats.npy'))
    #yTst = np.zeros((xTst.shape[0],), dtype=y.dtype)
    
    val = 1000
#     print (np.mean(x,1,keepdims=True))
    x, xTst = cifar_10_preprocess(x, xTst)    
 
    
    
    train_loader = DataLoader(TensorDataset(torch.from_numpy(x).float(), torch.from_numpy(y).long()),
                              batch_size = batch_size, shuffle=True)
    #val_loader = DataLoader(TensorDataset(torch.from_numpy(xval).float(), torch.from_numpy(yval).long()),
     #                       batch_size = batch_size, shuffle=True)
    test_loader = DataLoader(TensorDataset(torch.from_numpy(xTst).float(), torch.from_numpy(xTst).long()),
                            batch_size = 1, shuffle=False)
                
    return train_loader, train_loader , test_loader
```

```python
def main(argv):
    
    #initalize functions
    model = all_cnn_module()
    epochs = 20
    
    trainXt, valXt, testXt = myDataset()
    
    #define loss function values 
    #optimizer = optim.SGD(model.parameters(), lr=0.00001, momentum=0.9,weight_decay = 0.001)
    
    optimizer = optim.Adam(model.parameters(), lr=0.0001)
    
    criterion = nn.CrossEntropyLoss()
    
    if torch.cuda.is_available():
        model.cuda()
    
    #model.load_state_dict(torch.load("model.pt"))
    
    model.train()
    
    for epochs in range(epochs):
        
        # Train
        losses = []
        
        for (inputs, labels) in (trainXt):
            
            if torch.cuda.is_available():
                inputs = inputs.cuda()
                labels = labels.cuda()
                criterion = criterion.cuda()
                
                
            inputs = torch.autograd.Variable(inputs)
            labels = torch.autograd.Variable(labels)
            #print(labels)
            
            
            optimizer.zero_grad()
            
            #inputs = np.float32(inputs)
            
            #inputs = inputs.float32
            
            #print('Type', type(inputs))
    
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            losses.append(loss.cpu().detach().numpy())
            
        torch.save(model.state_dict(), 'model.pt')
            
        print(" the error ",np.mean(losses))
            
        print(" Training done")
        
            
    for epochs in range(2):
            
        for (inputs, labels) in (testXt):
            
            if torch.cuda.is_available():
                inputs = inputs.cuda()
                
            inputs = torch.autograd.Variable(inputs)
            
            pred  = model(inputs)
            #pred = pred.cpu()
            
        np.savetxt('submission.csv', pred, delimiter=',')
                
        write_results(model, pred)
```

```python
if __name__ == '__main__':
    main(sys.argv[1:])
```
